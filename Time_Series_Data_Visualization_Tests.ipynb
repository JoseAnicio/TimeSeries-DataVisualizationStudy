```python
# Import data and libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("fremont-bridge.csv", index_col='Date', parse_dates=True)
```

```python
# Adjust data and add columns
data.columns = ["West", "East"]
data['Total'] = data['West']+data['East']
```

```python
# Exploratory data visualizations
sns.set()
data.plot(style=[':', '--', '-'])
plt.ylabel("Hourly Bicycle count")
plt.show()
```

```python
weeklysum = data.resample("W").sum()
weeklysum.plot(style = [':', '--', '-'])
plt.ylabel('Weekly bicycle count')
plt.show()
```

```python
weeklymax = data.resample("W").max()
weeklymax.plot(style = [':', '--', '-'])
plt.ylabel('Weekly bicycle count')
plt.show()
```

```python
sns.set_context("paper")
sns.set_palette("deep")

dailysum = data.resample("D").sum()
dailysum.rolling(90, center = True).mean().plot(style = [':', '--', '-'])
plt.ylabel("Mean hourly count")
plt.show()
```

```python
monthsum = data.resample("ye").mean()
monthsum.rolling(2, center = True).mean().plot(kind='bar')
plt.ylabel("Mean hourly count")
plt.tight_layout()
plt.show()
```
